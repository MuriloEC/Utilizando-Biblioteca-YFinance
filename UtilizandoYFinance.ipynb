```python
import yfinance as yf
```

```python
petr = yf.download(["PETR4.SA"], period="30d")# 30d indica o periodo de 30 dias, outros periodos podem ser usados.

# Imprime os dados
print(petr)
```

```python
# Imprimindo apenas a coluna "Close"
print(petr["Close"])
```

```python
# Calculando a média móvel e adicionando os resultados em uma nova coluna da tabela
petr["MM30"] = petr["Close"].rolling(window=5).mean()
print(petr)
```

```python
import matplotlib.pyplot as plt
# Colocando no gráfico a variação do preço da ação nos 30 dias
petr['Close']['PETR4.SA'].plot()
plt.title("Histórico de preços da PETR4")
plt.show()
```

```python
import pandas as pd
import numpy as np

# Baixar dados da ação PETR4 (Petrobras) da B3
ticker = 'PETR4.SA'
dados = yf.download(ticker, start='2025-01-01', end='2025-05-31')

# Visualizar os 5 primeiros registros
print("Dados brutos:")
print(dados.head())

# Adicionar média móvel de 20 dias e 50 dias
dados['MM20'] = dados['Close'].rolling(window=20).mean()
dados['MM50'] = dados['Close'].rolling(window=50).mean()

# Calcular os retornos diários
dados['Retorno_Diario'] = dados['Close'].pct_change()

# Estatísticas com NumPy
media_retorno = np.mean(dados['Retorno_Diario'].dropna())
desvio_retorno = np.std(dados['Retorno_Diario'].dropna())
print(f"\nRetorno médio diário: {media_retorno:.5f}")
print(f"Desvio padrão do retorno diário: {desvio_retorno:.5f}")

# Plotar os preços de fechamento com médias móveis
plt.figure(figsize=(14, 6))
plt.plot(dados['Close'], label='Preço de Fechamento')
plt.plot(dados['MM20'], label='Média Móvel 20 dias')
plt.plot(dados['MM50'], label='Média Móvel 50 dias')
plt.title('PETR4 - Preço de Fechamento e Médias Móveis')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
